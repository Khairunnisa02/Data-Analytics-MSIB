```python

target_counts = df['TARGET'].value_counts()

# Mengatur ukuran gambar
plt.figure(figsize=(6, 6))

# Menggambar diagram pie
plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%')

# Memberikan judul diagram
plt.title('Persentase Kategori TARGET')

# Menampilkan diagram pie
plt.show()
```

>O = tidak gagal bayar,
>1 = gagal bayar

berdasarkan pie chart diatas,terdapat 8.1% peminjam yang gagal bayar

```python
# Membagi plot menjadi grid 2 baris 2 kolom
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Menggambar countplot di setiap subplot
sns.countplot(data=df, x="NAME_CONTRACT_TYPE", hue="TARGET", ax=axes[0, 0])
sns.countplot(data=df, x="CODE_GENDER", hue="TARGET", ax=axes[0, 1])
sns.countplot(data=df, x="FLAG_OWN_CAR", hue="TARGET", ax=axes[1, 0])
sns.countplot(data=df, x="FLAG_OWN_REALTY", hue="TARGET", ax=axes[1, 1])

# Menambahkan presentase di setiap bar
total_count = len(df)  # Total jumlah data dalam DataFrame

for ax in axes.flatten():
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height() / total_count)
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax.annotate(percentage, (x, y), ha='center', va='bottom')

# Menampilkan plot
plt.tight_layout()
plt.show()
```

Kesimpulan dari visualisasi diatas ialah
>Kolom 'CODE_GENDER' Peminjam terbanyak adalah perempuan dengan total 61.2% dan male 30.7% dan juga terdapat entri data "XNA" yang tidak menyediakan informasi apapun sehingga bisa dihapus.

>Kolom 'NAME_CONTRACT_TYPE" Kebanyakan orang mengambil pinjaman dalam bentuk pinjaman tunai daripada pinjaman bergulir seperti kartu kredit.

###  3.2.1 Univariate Analysis

### CREDIT_ACTIVE

```python
fig, axes = plt.subplots(1, 1, figsize=(10, 8))

ax = sns.countplot(data=df_b, y="CREDIT_ACTIVE", orient='h',
                   order=df_b["CREDIT_ACTIVE"].value_counts().index, ax=axes)

total = len(df_b["CREDIT_ACTIVE"])  # Total jumlah data

for p in ax.patches:
    width = p.get_width()
    ax.text(width + 3, p.get_y() + p.get_height() / 2, f"{width/total:.2%}", va="center")

ax.set_title("distribusi untuk Credit Active Status di Bureau Dataset")
ax.legend(loc='lower right')
plt.show()
```

>Active: Nilai "Active" menunjukkan bahwa kredit masih berlangsung dan belum ditutup. Ini berarti pelanggan masih memiliki kewajiban untuk membayar cicilan atau saldo yang terhutang pada kredit tersebut.

>Closed: Nilai "Closed" menunjukkan bahwa kredit tersebut sudah ditutup. Ini berarti pelanggan telah melunasi semua pembayaran yang terhutang dan tidak memiliki kewajiban pembayaran lebih lanjut terkait dengan kredit tersebut.

>Sold: Nilai "Sold" menunjukkan bahwa kredit telah dijual kepada pihak ketiga. Ini berarti lembaga keuangan awal yang memberikan kredit telah mentransfer hak tagihan kredit kepada pihak lain, seperti lembaga keuangan atau kolektor hutang. Pelanggan masih memiliki kewajiban untuk membayar hutang, tetapi kepada pihak yang membeli kredit.

>Bad debt: Nilai "Bad debt" menunjukkan bahwa kredit tersebut dianggap sebagai hutang macet. Ini berarti pelanggan tidak melakukan pembayaran yang diperlukan atau ada masalah lain yang menyebabkan kredit dianggap tidak terbayarkan.
